#### About Rows: 

Each row contains an ISBN, or collection of ISBNs, or book identifier(s). 

A row is generated for each ISBN / collection of ISBNs for a copy of a material for each format available if it is checked out in a given month.

So, if *Book Z* has three (3) checkouts in a month, one ebook, one physical book, and one audiobook, the book will generate three (3) rows with a checkout count of one (1) each in each row. Additionally, in some cases, if there is more than one keyed-in variant attached to a book, say there is both *Book Z* and *Book Z: Anniversary Copy* also in digital, physical, and audiobook, there are now six (6) rows for the book in one month if each version has at least one checkout.

#### Possible Limitations: 

Limitations are occurrences that could impact observations in a dataset, some of the ones SPL data may contain are: 

* SPL's physical or digital limitations on numbers of a particular copy of a material (SPL is unlikely to be able to loan a material 10+ times the same month if they have 1 copy).
* There are physical limitations, or lag / delay, to physical book borrowing that applies less to ebooks, in that, with a physical book, someone needs to bring the book back to the library, the librarian needs to log the book as returned, then contact the next person on the waitlist, and the next waitlisted reader has to come get the book (or some variation of those steps), whereas with a digital copy on Libby, once the ebook is returned, it goes right to the next person on the waitlist to borrow or skip, regardless of if it is during "operational hours" or 3 o'clock in the morning. 
* Likewise with some physical copies, some numbers might be impacted from COVID-19 lockdowns.  
* Numbers could be skewed because this is ***checkout information*** so if someone, say, borrows *Shadow and Bone* in 2013, but then re-borrows it in 2022, there will be two counts (2) for the book.
* Author or Creator attributions appear to be limited to one (1) author/creator per work, which creates blank or null spaces in the related columns for anthologies and complicates the parsing of other multi-author works.

I am unsure if Leigh Bardugo has ever participated in an anthology collection, and if so, if SPL would stock any related titles, but if she has, and it is a multi-author work, it may not be represented herein.
